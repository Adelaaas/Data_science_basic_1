```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression

from google.colab import files
uploaded = files.upload()

train = pd.read_csv("jobclassinfo2.csv")
train.head()
```

```python
print(train.info())
```

```python
train['JobFamily'].value_counts()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder() 
label_encoder.fit(train['JobFamilyDescription']) 
train['JobFamilyDescription'] = label_encoder.transform(train['JobFamilyDescription'])
label_encoder.fit(train['JobClassDescription']) 
train['JobClassDescription'] = label_encoder.transform(train['JobClassDescription'])
label_encoder.fit(train['PG']) 
train['PG'] = label_encoder.transform(train['PG'])

train.head()
```

```python
features = pd.Series(['JobFamilyDescription', 'JobClass', 'JobClassDescription', 'PayGrade', 'EducationLevel', 'Experience', 'OrgImpact', 'ProblemSolving', 'Supervision', 'ContactLevel', 'FinancialBudget'])
from sklearn.linear_model import LogisticRegression
X = train[['JobFamilyDescription', 'JobClass', 'JobClassDescription', 'PayGrade', 'EducationLevel', 'Experience', 'OrgImpact', 'ProblemSolving', 'Supervision', 'ContactLevel', 'FinancialBudget']].values
y = train['JobFamily'].values

slr = LogisticRegression() 
slr.fit(X, y)
coeff_df = pd.DataFrame(slr.coef_)  
coeff_df['features'] = features

coeff_df
```

```python
train = train.drop(['ID'], axis=1)
train = train.drop(['JobFamilyDescription'], axis=1)
train
```

```python
from sklearn.model_selection import train_test_split

# train - на тренировочные и тестовые
X = train.drop(['JobFamily'], axis=1)
y = train['JobFamily']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)

model = LogisticRegression()

model.fit(X_train, y_train)

answers_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

# answers_pred - ответы которые вернула модель для X_test
# y_test - это правильные ответы для X_test
print(accuracy_score(y_test,answers_pred))
```
