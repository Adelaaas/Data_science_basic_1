```python
train = pd.read_csv("Test.csv")
train
```

```python
print(train.shape)
print(train.info())
```

```python
train = train.dropna()
```

```python
print(train.isna().sum())
```

```python
train.columns
```

```python
train = pd.get_dummies(train, columns=['ID', 'Gender', 'Ever_Married', 'Age', 'Graduated', 'Profession',
       'Work_Experience', 'Spending_Score', 'Family_Size', 'Var_1',
       'Segmentation'])
train
```

```python
from sklearn.model_selection import train_test_split

# train - на тренировочные и тестовые
X = train.drop(['Segmentation_A'], axis=1)
y = train['Segmentation_A']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)

model = LogisticRegression()

model.fit(X_train, y_train)

answers_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test,answers_pred))
```

```python
from sklearn.linear_model import LinearRegression
x = train.values
y = train['Segmentation_A'].values

feachures = pd.Series(['ID', 'Gender', 'Ever_Married', 'Age', 'Graduated', 'Profession',
       'Work_Experience', 'Spending_Score', 'Family_Size', 'Var_1',
       'Segmentation'])
feachures
```

```python
linearRegression = LinearRegression()
linearRegression.fit(x, y)

coeff = pd.DataFrame(linearRegression.coef_, columns=['coefficient'])  
coeff['feachures'] = feachures

coeff
```

```python
print("-" * 30)
print(f"w0 = {linearRegression.intercept_}")
print("-" * 30)
print(f"w1 = {linearRegression.coef_}")
print("-" * 30)
print("y = w0 + w1 * x")
print(f"y = {linearRegression.intercept_} + {linearRegression.coef_} * x")
```
