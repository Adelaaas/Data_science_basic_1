```python
for i in range(len(nan_cols)):
    print(nan_cols[i])
    # print(train[nan_cols[i]].mean())
    train[nan_cols[i]].fillna(value = train[nan_cols[i]].mean(), inplace = True)
```

```python
train.isna().sum()
```

```python
train = train.drop(columns = ['WindGustDir', 'WindDir9am', 'WindDir3pm'])
train.isna().sum()
```

```python
from sklearn import preprocessing 

label_encoder = preprocessing.LabelEncoder() 

label_encoder.fit(train['Location']) 
train['Location'] = label_encoder.transform(train['Location']) 
train
```

```python
from sklearn.linear_model import LogisticRegression # логистическая регрессия

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix

X = train.drop(columns=['RainTomorrow'])
y = train['RainTomorrow']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=42)

log = LogisticRegression()

log.fit(X_train, y_train)

# predict - тестирование
y_pred = log.predict(X_test)

# оценка классификатор/модели
score = accuracy_score(y_test, y_pred)
```
