```python
from sklearn import preprocessing

model = preprocessing.LabelEncoder()

model.fit(df['class'])
df['class'] = model.transform(df['class'])

model.fit(df['cap-shape'])
df['cap-shape'] = model.transform(df['cap-shape'])

model.fit(df['cap-surface'])
df['cap-surface'] = model.transform(df['cap-surface'])

model.fit(df['cap-color'])
df['cap-color'] = model.transform(df['cap-color'])

model.fit(df['bruises'])
df['bruises'] = model.transform(df['bruises'])

model.fit(df['odor'])
df['odor'] = model.transform(df['odor'])

model.fit(df['gill-attachment'])
df['gill-attachment'] = model.transform(df['gill-attachment'])

model.fit(df['gill-spacing'])
df['gill-spacing'] = model.transform(df['gill-spacing'])

model.fit(df['gill-size'])
df['gill-size'] = model.transform(df['gill-size'])

model.fit(df['gill-color'])
df['gill-color'] = model.transform(df['gill-color'])

model.fit(df['stalk-shape'])
df['stalk-shape'] = model.transform(df['stalk-shape'])

model.fit(df['stalk-root'])
df['stalk-root'] = model.transform(df['stalk-root'])

model.fit(df['stalk-surface-above-ring'])
df['stalk-surface-above-ring'] = model.transform(df['stalk-surface-above-ring'])

model.fit(df['stalk-surface-below-ring'])
df['stalk-surface-below-ring'] = model.transform(df['stalk-surface-below-ring'])

model.fit(df['stalk-color-above-ring'])
df['stalk-color-above-ring'] = model.transform(df['stalk-color-above-ring'])

model.fit(df['stalk-color-below-ring'])
df['stalk-color-below-ring'] = model.transform(df['stalk-color-below-ring'])

model.fit(df['veil-type'])
df['veil-type'] = model.transform(df['veil-type'])

model.fit(df['veil-color'])
df['veil-color'] = model.transform(df['veil-color'])

model.fit(df['ring-number'])
df['ring-number'] = model.transform(df['ring-number'])

model.fit(df['ring-type'])
df['ring-type'] = model.transform(df['ring-type'])

model.fit(df['spore-print-color'])
df['spore-print-color'] = model.transform(df['spore-print-color'])

model.fit(df['population'])
df['population'] = model.transform(df['population'])

model.fit(df['habitat'])
df['habitat'] = model.transform(df['habitat'])

df
```

```python
x = df.drop(['class'], axis=1)
y = df['class']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1/3, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

scores = []
for i in range(1, 20):
    knn = KNeighborsClassifier(i)
    knn.fit(x_train, y_train)
    answers = knn.predict(x_test)
    scores.append(knn.score(x_test, y_test))
print(scores)
```

```python
max_score = max(scores)
ind = [i for i, v in enumerate(scores) if v == max_score]
ind
```

```python
knn = KNeighborsClassifier(1)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import recall_score, precision_score

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("precision:", precision_score(y_test, y_pred))
```
